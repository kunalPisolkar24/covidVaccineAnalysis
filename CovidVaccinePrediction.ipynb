```python
import pandas as pd
df = pd.read_csv("/content/covid_vaccine_statewise.csv")
print(df.dtypes)
```

```python
import numpy as np
states = np.unique(df["State"])
mh = df[df["State"]=="Maharashtra"]
print(mh["First Dose Administered"].sum())
```

```python
for state in states:
  temp = df[df["State"] == state]
  print(state, " --> ", temp["First Dose Administered"].sum())
```

```python
first_dose = df[(df["Updated On"] == "09/08/2021") & (df["State"] != "India")]
```
